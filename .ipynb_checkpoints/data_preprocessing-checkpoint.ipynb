```python
import pandas as pd
import os 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import joblib
```

```python
training_ds_path = 'Training_Data'
concatenated_data_file = 'concat_data.csv'

def preprocess_columns(df):
    feature_names = list(df.columns)
    final_feature_names = []
    for feature_name in feature_names:
        if ' ' in feature_name:
            feature_name = feature_name.replace(' ','_')
        final_feature_names.append(feature_name)
    return final_feature_names


def combine(training_ds_path):
    dfs = []
    filenames = os.listdir(training_ds_path)
    for filename in filenames:
        df = pd.read_csv(f'{training_ds_path}/{filename}')
        dfs.append(df)
    if concatenated_data_file in os.listdir():
        os.remove(concatenated_data_file)
    concat_df = pd.concat([dfs[i] for i in range(len(dfs))])
    concat_df.columns = preprocess_columns(concat_df)
    concat_df.to_csv('concat_data.csv', index = False)
```

```python
combine(training_ds_path)
df = pd.read_csv('concat_data.csv')
df.head()
```

```python
df.columns = [i.lower() for i in df.columns]
selected_features = ['limit_bal','sex','education','marriage','age','pay_amt6']
X = df[selected_features]
y = df['default_payment_next_month']
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1,test_size=0.2)
print('X_train shape :',X_train.shape)
print('y_train shape :',y_train.shape)
print('X_test shape :',X_test.shape)
print('y_test shape :',y_test.shape)
```

```python
# saving the training and testing data as a dataframe 
training_df = pd.concat([X_train,y_train],axis = 1)
validation_df = pd.concat([X_test,y_test],axis = 1)
training_df.to_csv('training_data.csv',index = False)
validation_df.to_csv('validation_data.csv',index = False)
```

```python
# random forest model 
model = RandomForestClassifier().fit(X_train,y_train)
model_score = model.score(X_test,y_test)

lrmodel = LogisticRegression().fit(X_train,y_train)
print('logistic regression score = ',lrmodel.score(X_test,y_test))

adamodel = AdaBoostClassifier().fit(X_train,y_train)
print('Adaboost Classifier score = ',adamodel.score(X_test,y_test))

grbmodel = GradientBoostingClassifier().fit(X_train,y_train)
print('gradient Classifier score = ',grbmodel.score(X_test,y_test))



print('random forest model_score = ',model_score)
```

```python
model_names = ['randomForest','gradientClassifier','adaBoost','logisticRegressor']
model_scores = [0.98,0.82,0.799,0.795]
from matplotlib  import pyplot as plt
plt.bar(x = model_names, height = model_scores)
plt.ylim(0.7, 1)

plt.title('Model Comparison')

plt.xticks(rotation = '30')
plt.ylabel('Model scores')
plt.show()
```

```python
predictions = model.predict(X_test)
actual = y_test
model_evaluation = pd.DataFrame()
model_evaluation['actual'] = y_test
model_evaluation['predictions'] = predictions
model_evaluation
```
